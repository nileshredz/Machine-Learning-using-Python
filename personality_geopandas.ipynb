```python
pos_q=['EXT1','EXT3','EXT5','EXT7','EXT9', #Extroversion.
      'EST1','EST3','EST5','EST6','EST7','EST8','EST9','EST10', #Neuroticism
      'AGR2','AGR4','AGR6','AGR8','AGR9','AGR10', #Agreeableness
      'CSN1','CSN3','CSN5','CSN7','CSN9','CSN10', #Conscientiousness
      'OPN1','OPN3','OPN5','OPN7','OPN8','OPN9','OPN10']#Openess
```

```python
neg_q=['EXT2','EXT4','EXT6','EXT8','EXT10', #Extroversion
    'EST2','EST4',  #Neuroticism
    'AGR1','AGR3','AGR5','AGR7',#Agreeableness
    'CSN2','CSN4','CSN6','CSN8',#Conscientiousness
    'OPN2','OPN4','OPN6'#Openness
]
```

```python

df.EXT1.value_counts()
```

```python
df[pos_q] = df[pos_q].replace({1:-2, 2:-1, 3:0, 4:1, 5:2})
df[neg_q] = df[neg_q].replace({1:2, 2:1, 3:0, 4:-1, 5:-2})
```

```python
cols=pos_q+neg_q
df=df[sorted(cols)]
df.head()
```

```python
chars=['EXT', 'EST','AGR','CSN','OPN']
chars_=['Extroversion','Neuroticism', 'Agreeableness','Conscientiousness','Openness']
```

```python
for char in chars:
    char_c=sorted([i for i in df.columns if
                  char in i and'_E' not in i])
    df[char]=df[char_c].sum(axis=1)
df[chars].head()
```

```python
fig, axs=plt.subplots(ncols=2, nrows=3,figsize=(18,9))
plt.subplots_adjust(left=None,bottom=None,right=None,top=1.3, wspace=None, hspace=None)
row = -1;col = 2
for i, (char,char_) in enumerate (zip(chars,chars_)):
    if not i%2:
        row +=1
    if not i%2:
        col -=2
    i +=col
    sns.distplot(df[char], ax=axs[row][i],axlabel ='', kde=False, bins=40).set_title(char_,pad=10)
fig.delaxes(axs[2][1])
```
